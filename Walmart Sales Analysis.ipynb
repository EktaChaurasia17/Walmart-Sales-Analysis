## City Wise Profit

```python
profit_by_city = df.groupby(['City'], as_index = False)['Profit'].sum().head(10)

ax=sns.barplot( data =profit_by_city  , x = 'Profit' , y = 'City');
for bars in ax.containers:
    ax.bar_label(bars);
```

## Category Wise Profit

```python
profit_by_city = df.groupby(['Category'], as_index = False)['Profit'].sum().head(10)

ax=sns.barplot( data =profit_by_city  , x = 'Profit' , y = 'Category')
for bars in ax.containers:
    ax.bar_label(bars)
```
